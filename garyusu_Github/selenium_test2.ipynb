```python
#Webdrvier 位址
webdriver_path = "chromedriver.exe"
#打開模擬瀏覽器(無參數則開啟同目錄下的)
driver = webdriver.Chrome()
#前往網頁
driver.get("https://fanyi.baidu.com/#jp/cht")
```

```python

input_jp_list=[]
# output_cht_list=[]
file_path = "[日]呪と罪と流転の華達　～最強賢者は最愛の恋人を寝取らせる～/第※話　とある一日[作者：流星].txt"

with open(file_path, "r", encoding="utf-8") as file:
    #過濾換行符號and全形空白 加到 input_jp_list
    for f in file.readlines():
        if f!= '\n' and f!= '\u3000\n':
            input_jp_list.append(f)

n = 0 
temp_list = ""
input_list = []
with open("test.txt", "a", encoding="utf-8") as file2:

    #每n個字串串接後加到 ouput_list
    for input_jp in input_jp_list: 
        if n == 50:
            print(temp_list)
            input_list.append(temp_list)
            temp_list ="" #字串清空
            n =0 #重新計算
        #將字串串接起來
        temp_list+=input_jp
        n+=1 #計算迴圈次數

    #開始丟到區塊翻譯
    for inp in input_list:
        
        #一次n行輸入欲翻譯區塊
        key_in = driver.find_element_by_xpath('//*[@id="baidu_translate_input"]').send_keys(inp)
        #等待翻譯時間 設定秒數
        time.sleep(5)
        #找到輸出區塊內的文字
        output_cht = driver.find_element_by_css_selector('p.ordinary-output.target-output.clearfix').text
        #寫入檔案並加上換行
        file2.write('\n' + input_jp + output_cht + '\n')
        print( input_jp + output_cht )
        #清除框內文字
        key_input = driver.find_element_by_xpath('//*[@id="baidu_translate_input"]').clear()
    
    
        
```
